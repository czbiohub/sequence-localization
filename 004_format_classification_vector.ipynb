```python
txt = os.path.join(input_folder, 'antibody_celline_location_v17.txt')
localization = pd.read_table(txt)
print(localization.shape)
localization.head()
```

```python
nfc1 = localization['Nucleoli_fibrillar_center']
nfc2 = localization['Nucleoli_fibrillar_center.1']


(nfc1 == nfc2).all()
```

Second "Nucleoli_fibrillar_center" is exactly the same, remove.

```python
print(localization.shape)
localization_v2 = localization.drop('Nucleoli_fibrillar_center.1', axis=1)
print(localization_v2.shape)
localization_v2.head()
```

## What we want

- `tissue_name x ensembl_id x subcellular_localization` matrix
- Want this for each of Levels A, B and C for the last dimension, `subcellular_localization`. i.e. for level C, there are only 6 columns but for level A there are 29

```python
sizes = localization_v2.groupby(('ensg_id', 'tissue_name')).size()
sizes[sizes != 1]
```

```python
localization_v2.query('ensg_id == "ENSG00000262660,ENSG00000262814"')
```

So there's multiple antibodies per ensembl id.. Sounds like the way to go is actually antibody ids

## What we want, v2

- `tissue_name x antibody_id x subcellular_localization` matrix
- Want this for each of Levels A, B and C for the last dimension, `subcellular_localization`. i.e. for level C, there are only 6 columns but for level A there are 29

```python
sizes = localization_v2.groupby(('antibody_id', 'tissue_name')).size()
sizes[sizes != 1]
```

Yep, that's completely unique.

```python
columns_to_drop = ['ensg_id', 'gene_name', 'uniprot_id', 'prest_id', 'aa_sequencing_sequence']
localization_v3 = localization_v2.drop(columns=columns_to_drop, axis=1)
localization_v3 = localization_v3.set_index(['antibody_id', 'tissue_name'])
print(localization_v3.shape)
localization_v3.head()
```
