```python
for cell_line, df in metadata.groupby('cell_line'):
    print(cell_line, df.shape,)
    read1 = df.acc + "_pass_1.fastq.gz"
    read2 = df.acc + "_pass_2.fastq.gz"
    reads = ' '.join(read1) + ' ' + ' '.join(read2)
    
    # Sanitize the cell line name
    cell_line = cell_line.replace(' ', '_')
    output_folder = f"/mnt/data/khmer_k32/{cell_line}"
    output_file = f'{output_folder}/khmer_countgraph'
#     print(output_file)
    ! mkdir -p $output_folder
    # ksize of 32 is the default
    ! load-into-counting.py --summary-info tsv --max-memory-usage 60G --ksize 32 --threads 16  $output_file $reads
    
```
