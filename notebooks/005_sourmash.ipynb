## Calculate RNA signatures

```python
for cell_line, df in metadata.groupby('cell_line'):
    print('----', cell_line, df.shape, '----')
    # Sanitize the cell line name
    cell_line = cell_line.replace(' ', '_')

    # Use previously calculated countgraphs
    read1 = df.acc + "_pass_1.fastq.gz"
    read2 = df.acc + "_pass_2.fastq.gz"
    reads = ' '.join(read1) + ' ' + ' '.join(read2)
    khmer_folder = f"/mnt/data/khmer_k32/{cell_line}"
    khmer_countgraph = f'{khmer_folder}/khmer_countgraph'
    ! trim-low-abund.py --cutoff 3 --trim-at-coverage 18 --variable-coverage --max-memory-usage 64e9 --loadgraph $khmer_countgraph $reads
        
    trimmed_read1 = df.acc + "_pass_1.fastq.gz.abundtrim"
    trimmed_read2 = df.acc + "_pass_2.fastq.gz.abundtrim"
    trimmed_reads = ' '.join(trimmed_read1) + ' ' + ' '.join(trimmed_read2)
    
    output_folder = f"/mnt/data/sourmash/{cell_line}"
    output_file = f'{output_folder}/sourmash_rna_signature'
    

#     print(output_file)
    ! mkdir -p $output_folder
    # ksize of 32 is the default
    ! sourmash compute --track-abundance --merge $cell_line --output $output_file $trimmed_reads
    
```

## Calculate protein signatures

Also calculate protein signatures, but ksizes must be divisible by 3

```python
for cell_line, df in metadata.groupby('cell_line'):
    print('----', cell_line, df.shape, '----')
    read1 = df.acc + "_pass_1.fastq.gz"
    read2 = df.acc + "_pass_2.fastq.gz"
    reads = ' '.join(read1) + ' ' + ' '.join(read2)
    
    # Sanitize the cell line name
    cell_line = cell_line.replace(' ', '_')
    output_folder = f"/mnt/data/sourmash/{cell_line}"
    output_file = f'{output_folder}/sourmash_protein_signature'
#     print(output_file)
    ! mkdir -p $output_folder
    # ksize of 32 is the default
    ! sourmash compute --track-abundance --merge $cell_line --output $output_file --ksizes 21, 33, 51 --protein $reads
    
```
