### Suggested fastq-dump parameters from blog post


From: https://edwards.sdsu.edu/research/fastq-dump/

````python
%%bash
# Extract SRA files to fastq in parallel
export CORES=16

export FASTQ_DUMP_FLAGS='--outdir /mnt/data/fastq_dump_v1 --gzip --skip-technical  --readids --read-filter pass --dumpbase --split-3 --clip'
ls -1  /mnt/data/rawdata/*/*.sra \
    | xargs -P ${CORES} -I{} bash -c "fastq-dump $FASTQ_DUMP_FLAGS {}"
````
