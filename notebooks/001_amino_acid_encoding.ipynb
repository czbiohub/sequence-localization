```python
import pandas as pd
from pandas.compat import StringIO

aa = pd.read_table(StringIO(s))
# aa = pd.read_excel('/Users/olgabot/Documents/amino_acids.xlsx')
aa = aa.dropna(how='all')
aa = aa.set_index("1-letter code")

# Drop double row for histidine
aa
```

```python
pd.options.display.max_columns = 50
```

```python
aa_one_hot_encoding = pd.get_dummies(aa.index)
aa_one_hot_encoding = aa_one_hot_encoding.set_index(aa.index)
print(aa_one_hot_encoding.shape)
aa_one_hot_encoding
```

```python
side_chain_class = pd.get_dummies(aa['Side chain class'])
side_chain_class
```

```python
polarity = pd.get_dummies(aa['Side chain polarity'])
polarity
```

```python
charge = pd.get_dummies(aa['Side chain charge'])
charge = charge.drop(columns='positive(10%)/neutral(90%)')

# Histidine is a special case
charge.loc['H', 'neutral'] = 0.9
charge.loc['H', 'positive'] = 0.1

charge
```

## Write encodings to CSV

```python
dfs = {'one_hot_encoding': aa_one_hot_encoding, 
       'side_chain_polarity': polarity, "side_chain_charge": charge, 
       'side_chain_class': side_chain_class, 'molecular_weight': aa[['MW']]}

for name, df in dfs.items():
    csv = os.path.join(data_folder, name + '.csv')
    df.to_csv(csv)
```
