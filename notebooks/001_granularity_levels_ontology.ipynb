```python
from pandas.compat import StringIO
aa = pd.read_excel('/Users/olgabot/Documents/amino_acids.xlsx')
aa = aa.dropna(how='all')
aa
```

```python
aa.dtypes
```

```python
aa['Hydropathy index'].str.replace('−', '-')
```

```python
aa['Hydropathy index'].map(type)
```

```python
aa['Hydropathy index'] = aa['Hydropathy index'].map(lambda x: float(x.replace(u'\u2212', '-')) if not isinstance(x, float) else x)
```

```python
aa.groupby('Side chain class').size()
```

```python
aa.groupby('Side chain polarity').size()
```

```python
aa.groupby(['Side chain polarity', 'Side chain class']).size()
```

```python
aa.groupby(['Side chain polarity', 'Side chain class', 'Side chain charge']).size()
```

```python
aa.groupby('Side chain charge').size()
```

```python
bin(2)
```

```python
aa['is_polar'] = aa['Side chain polarity'] != 'nonpolar'
aa.head()
```

```python
aa['simple_binary'] = aa.index.map(lambda x: '{:05b}'.format(x))
aa.head()
```

```python
last_four_bits = aa.groupby('is_polar', group_keys=False).apply(
    lambda x: pd.Series(map(lambda y: '{:04b}'.format(y), range(len(x))),
                            index=x.index))
last_four_bits
```

```python
first_bit = aa['is_polar'].astype(int).astype(str)
first_bit
```

```python
first_bit + last_four_bits
```

```python
aa['clever_binary'] = (first_bit + last_four_bits).astype(str)
aa['clever_binary'].head()
```

```python
aa.dtypes
```

```python
for x in aa['clever_binary']:
    print(x)
```

```python
aa['simple_binary_protected'] = "'" + aa['simple_binary']
aa['clever_binary_protected'] = "'" + aa['clever_binary']
```

```python
aa.to_csv('/Users/olgabot/code/sequence-localization/data/00_original/amino_acids.csv', #encoding='utf8', 
          index=False)
```
